```python
plt.figure(figsize=(15,3))
sns.kdeplot(year15["Happiness Score"], color="darkturquoise", shade=False)
sns.kdeplot(year16["Happiness Score"], color="lightcoral", shade=False)
sns.kdeplot(year17["Happiness.Score"], color="yellow", shade=False)
plt.legend(['2015', '2016', '2017'])
plt.title('Happiness Score Distribution of Countries')
plt.show()
```
