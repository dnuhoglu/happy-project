```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st

%matplotlib inline
```

```python
year15 = pd.read_csv('2015.csv')
year16 = pd.read_csv('2016.csv')
year17 = pd.read_csv('2017.csv')
```

```python
plt.figure(figsize=(15,3))
sns.kdeplot(year15["Economy (GDP per Capita)"], color="darkturquoise", shade=False)
sns.kdeplot(year16["Economy (GDP per Capita)"], color="lightcoral", shade=False)
sns.kdeplot(year17["Economy..GDP.per.Capita."], color="yellow", shade=False)
plt.legend(['2015', '2016', '2017'])
plt.title('Economy Score Distribution of Countries')
plt.show()
```

```python
year15['zscore'] = (year15["Economy (GDP per Capita)"] - year15["Economy (GDP per Capita)"].mean())
/ year15["Economy (GDP per Capita)"].std(ddof=0)
```

```python
year15.loc[year15['Country'] == 'Turkey'].zscore
```
